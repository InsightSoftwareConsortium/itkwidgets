```python
import pyvista
from pyvista import examples

from itkwidgets import view
```

```python
lidar = examples.download_lidar()

clipped = lidar.clip(origin=(0, 0, 1.76e3), normal=(0, 0, 1))
```

```python
view(point_set=clipped)
```
